```python
import logfire

from rich.pretty import pprint

from llm_agents.agents import GenericAssistant
from llm_agents.message_history import MongoDBMessageHistory
```

```python
logfire.configure(service_name="llm-agents")
_ = logfire.instrument_pydantic_ai()
```

```python
mongodb_message_history = MongoDBMessageHistory(
    session_id="0",
    message_limit=10,
)

generic_assitant = GenericAssistant(
    mongodb_message_history=mongodb_message_history,
)
```

```python
user_prompt_1 = "What is acid rain?"
agent_output_1 = await generic_assitant.generate(user_prompt=user_prompt_1)

pprint(agent_output_1)
```

```python
user_prompt_2 = "And why it is harmful to humans?"
agent_output_2 = await generic_assitant.generate(user_prompt=user_prompt_2)

pprint(agent_output_2)
```

```python
mongodb_message_history = generic_assitant.mongodb_message_history
assert mongodb_message_history is not None

message_history = await mongodb_message_history.get_messages()
pprint(message_history)
```

```python
user_prompt_3 = "Is also harmful to plants?"
agent_output_3 = await generic_assitant.generate(user_prompt=user_prompt_3)

pprint(agent_output_3)
```

```python
message_history = await mongodb_message_history.get_messages()
pprint(message_history)
```
